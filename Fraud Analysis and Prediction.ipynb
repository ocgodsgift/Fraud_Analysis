10. **Prediction and Modeling:**

    - What features are most important in predicting fraud?

```python
df_features = df[['AcountNumber', 'CVV', 'CustomerAge', 'Gender', 'Marital Status', 'CardType', 'TransactionType', 'Domain', 'Amount', 'NewBalance', 'OldBalance', 'Outcome']]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Create a LabelEncoder for each categorical column
label_encoders = {}
categorical_columns = ["Gender", "Marital Status", "CardType", "TransactionType", "Domain"]
```

```python
le = LabelEncoder()

for column in categorical_columns:
    df_features[column] = le.fit_transform(df_features[column])
```

```python
df_features
```

    - Can you build a predictive model to identify potential fraud cases?

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X = df_features.iloc[:,:-1].values
y = df_features.iloc[:,-1].values
```

```python
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
model = rfc.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, accuracy_score, confusion_matrix, classification_report, roc_curve, auc
```

```python
confusion_matrix = confusion_matrix(y_test, y_pred)
accuracy_score = accuracy_score(y_test, y_pred)
classification_report = classification_report(y_test, y_pred)
```

```python
ac = np.round(accuracy_score * 100,2)
print(f'Accuracy Score: {ac}%')
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(confusion_matrix, cmap='inferno', cbar=False, annot=True, fmt='d')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
print(f'Classification Report \n {classification_report}')
```

```python
from sklearn.model_selection import cross_val_score
```

```python
# Perform 5-fold cross-validation and calculate accuracy
scores = cross_val_score(rfc, X, y, cv=3, scoring='accuracy')

# Print the cross-validation scores
print("Cross-validation scores:", scores)
print("Mean accuracy:", scores.mean())
```

```python
# Plot the cross-validation scores
plt.figure(figsize=(6,5))
plt.bar(range(1, len(scores) + 1), scores)
plt.xlabel('Fold')
plt.ylabel('Accuracy')
plt.title('Cross-Validation Scores')
plt.show()
```

```python
mse = mean_squared_error(y_test,y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
# Predict probabilities of class 1 (positive class) on the test set
y_prob = model.predict_proba(X_test)[:, 1]
```

```python
# Calculate ROC curve and ROC AUC
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)
```

```python
# Plot ROC curve
plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, color='darkorange', lw=1, label=f'ROC curve (AUC = {np.round(roc_auc,2)})')
plt.plot([0, 1], [0, 1], color='navy', lw=1, linestyle='dotted')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('Receiver Operating Characteristic (ROC) Curve', fontsize=10)
plt.legend(loc='lower right')
plt.show()
```
